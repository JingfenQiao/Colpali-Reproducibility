```python
import json
import numpy as np
from collections import defaultdict

def statistical_measure(input_list):
    mean = np.mean(input_list)
    max = np.max(input_list)
    min = np.min(input_list)
    # print(f"Mean : {mean:.2f}, Max : {max:.2f}, Min : {min:.2f}")
    # return {f"Mean : {mean:.2f}, Max : {max:.2f}, Min : {min:.2f}"}
    return mean

def collect_converage(input_path):
    text_coverages = [] 
    with open(input_path, "r") as infn:
        for line in infn:
            data = json.loads(line)
            text_coverages.append(data['text_coverage'])

    text_coverages = np.array(text_coverages)
    return text_coverages

def collect_salience(input_path):
    text_salience = []
    with open(input_path, "r") as infn:
        for line in infn:
            data = json.loads(line)
            text_salience.append(data['text_salience'])

    text_salience = np.array(text_salience)
    return text_salience

def collect_estimation(input_path):
    queries = defaultdict(dict)
    with open(input_path, "r") as infn:
        for line in infn:
            data = json.loads(line)
            queries[data['query']] = data
    return queries

# def collect_queries(input_path, value):
#     queries =  []
#     with open(input_path, "r") as infn:
#         data = json.load(infn)
#         metric = "ndcg_cut_" + value
#         for key, val in data.items():
#             if val[metric] == 1:
#                 queries.append(key)
#     return queries

def collect_queries(input_path, value):
    queries =  []
    with open(input_path, "r") as infn:
        data = json.load(infn)
        for key, val in data.items():
            if val["ndcg_cut_10"] == value:
                queries.append(key)
    return queries
```

```python
datasets = ["arxivqa_test_subsampled", "docvqa_test_subsampled", "infovqa_test_subsampled", "tabfquad_test_subsampled", "tatdqa_test", "shiftproject_test", "syntheticDocQA_artificial_intelligence_test", "syntheticDocQA_energy_test", "syntheticDocQA_government_reports_test", "syntheticDocQA_healthcare_industry_test"]
estimation = defaultdict(dict)
for name in datasets:
    input_path = f"/ivi/ilps/personal/jqiao/colpali/outputs/converage/{name}.jsonl"
    estimation.update(collect_estimation(input_path))
```
